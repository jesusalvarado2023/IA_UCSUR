# Desarrollo de mi primer modelo de IA

```python
X = df[iris.feature_names]
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier

modelo = RandomForestClassifier(random_state=100)
modelo.fit(X_train, y_train)
```

¿Probemos un valor?

```python
nueva_entrada_df = pd.DataFrame([{
    'sepal length (cm)': 5.1,
    'sepal width (cm)': 3.5,
    'petal length (cm)': 1.4,
    'petal width (cm)': 0.2
}])

# Predicción:
y_pred1 = modelo.predict(nueva_entrada_df)
y_pred1
```

# 3. Métricas

Insertando el 20% que no se usó para entrenar:

```python
y_pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import classification_report

reporte = classification_report(y_test, y_pred, target_names=iris.target_names)
print(reporte)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap='Blues')
plt.title("Matriz de Confusión")
plt.show()
```
